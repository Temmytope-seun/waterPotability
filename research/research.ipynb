```python
## explanation ffor ensure annotation

def get_product(x:int,y:int) -> int:
    return x*y
```

```python
get_product(2,2)
```

```python
get_product(2,'4')
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_product(x:int,y:int) -> int:
    return x*y
```

```python
get_product(2,'4')
```
